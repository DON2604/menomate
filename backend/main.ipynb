```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score

datas = pd.read_csv('./tips.csv')

le = LabelEncoder()

y=datas[['tips']]
y.tips=le.fit_transform(y.tips)
y
```

```python
x=datas[["days_passed","cycle_day","flow","mood","hormonal_phase"]]
x.days_passed=le.fit_transform(x.days_passed)
x.cycle_days=le.fit_transform(x.cycle_day)
x.flow=le.fit_transform(x.flow)
x.mood=le.fit_transform(x.mood)
x.hormonal_phase=le.fit_transform(x.hormonal_phase)
x

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=15)
```

```python
model = RandomForestClassifier(random_state=30)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy",accuracy*100)
```
